```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
X = pd.read_csv('A2Q3.csv', header=None)
X = X.values
#Try different values of k and calculate the Silhouette score and BIC and log-likelihood 
for i in range(2,7):
    gmm = GaussianMixture(n_components=i, covariance_type='full', random_state=0)
    gmm.fit(X)
    labels = gmm.predict(X)
    #print(labels)
    #print(X)
    print('Silhouette score for k =',i,'is',silhouette_score(X, labels))
    print('BIC for k =',i,'is',gmm.bic(X))
    print('Log-likelihood for k =',i,'is',gmm.score(X))
    print('---------------------------------------------')
```

```python
#find the best k using the BIC and silhouette score
bic = []
sil = []
for i in range(2,10):
    gmm = GaussianMixture(n_components=i, covariance_type='full', random_state=0)
    gmm.fit(X)
    labels = gmm.predict(X)
    bic.append(gmm.bic(X))
    sil.append(silhouette_score(X, labels))
plt.plot(range(2,10), bic, 'bx-')
plt.xlabel('k')
plt.ylabel('BIC')
plt.title('The BIC for different values of k')
plt.show()
plt.plot(range(2,10), sil, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette score')
plt.title('The Silhouette score for different values of k')
plt.show()
```
